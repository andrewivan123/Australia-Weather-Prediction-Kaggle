```python
import pandas as pd
import numpy as np
```

```python
x=[]
for i in range(0, 6):
    p=pd.read_csv("DataReg/tab{}.csv".format(i))
    q=p[["RainToday", "Autumn", "Summer", "Winter", "Spring"]]
    x.append(q)
```

```python
x[0]
```

```python
for i in range(0,6):
    temp= x[i].drop(["RainToday"], axis=1)
    l=[]
    for k in range(0,len(temp)):
        for j in temp.columns:
            if(temp.at[k,j]==1):
                l.append(str(j))
    x[i]["Season"]=pd.Series(l)
```

```python
y=[]
for i in range(0,6):
    y.append(pd.DataFrame(x[i][["RainToday","Season"]]))
```

```python
y[0]
```

```python
dic={0: "New South Wales",
     1: "Victoria",
     2: "Queensland",
     3: "South Australia",
     4: "Western Australia",
     5: "Northern Territory"
}
```

```python
summer= pd.DataFrame(columns=["RainToday", "Region"])
spring= pd.DataFrame(columns=["RainToday", "Region"])
autumn= pd.DataFrame(columns=["RainToday", "Region"])
winter= pd.DataFrame(columns=["RainToday", "Region"])
for i in range(0,6):
    s= pd.DataFrame(y[i][y[i]["Season"]=="Summer"].drop(["Season"], axis=1))
    p= pd.DataFrame(y[i][y[i]["Season"]=="Spring"].drop(["Season"], axis=1))
    a= pd.DataFrame(y[i][y[i]["Season"]=="Autumn"].drop(["Season"], axis=1))
    w= pd.DataFrame(y[i][y[i]["Season"]=="Winter"].drop(["Season"], axis=1))
    s["Region"]=dic[i]
    p["Region"]=dic[i]
    a["Region"]=dic[i]
    w["Region"]=dic[i]
    summer = summer.append(s, ignore_index=True)
    spring = spring.append(p, ignore_index=True)
    autumn = autumn.append(a, ignore_index=True)
    winter = winter.append(w, ignore_index=True)
```

```python
summer
```

```python
summer.to_csv("summer.csv")
spring.to_csv("spring.csv")
autumn.to_csv("autumn.csv")
winter.to_csv("winter.csv")
```
