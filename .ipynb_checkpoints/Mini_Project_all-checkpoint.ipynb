```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
import pickle
sb.set()
```

# Data Cleaning and Preprocessing

Dropping rows with NaN values

```python
data_clean = data.dropna()
```

```python
data_clean.head()
```

## Grouping cities based on their state

```python
data_clean.Location.unique()
```

Mapping each city to their respective state

```python
reg= data_clean["Location"].drop_duplicates()
reg=reg.tolist()
toregion={}
for i in range(0,len(reg)):
    if i==4 or i==23:
        toregion[reg[i]]="DROPPED"
    elif i<=7:
        toregion[reg[i]]="NSW"
    elif i<=14:
        toregion[reg[i]]="VIC"
    elif i<=17:
        toregion[reg[i]]="QLD"
    elif i<=20:
        toregion[reg[i]]="SA"
    elif i<=22:
        toregion[reg[i]]="WA"
    else:
        toregion[reg[i]]="NT"
        
```

```python
toregion
```

Adding a new feature column called "Region" which represents the city's state

```python
data_clean["Region"]= data_clean["Location"].map(toregion)
```

```python
data_clean.head()
```

```python
f,axes = plt.subplots(1,1,figsize=(20,10))
sb.countplot(y=data_clean["Region"],)
```

Drop wind direction, location and risk_mm

```python
data_clean = data_clean.drop(["RISK_MM","Location","WindGustDir","WindDir9am","WindDir3pm"],axis = 1)
```

```python
data_clean.head()
```

```python
p = data_clean["Region"].unique()
p = p.tolist()
```

```python
data_split=[]
for i in range(0,len(p)):
    data_split.append(data_clean[data_clean["Region"]==p[i]])
```

Delete "DROPPED" Region

```python
del data_split[1]
```

## Converting dates into seasons

Convert string in "Date" to date-time data type

```python
for i in range(0, len(data_split)):
    data_split[i].sort_values(by="Date")
    data_split[i]=data_split[i].reset_index(drop=True)
    for j in range(0, data_split[i].shape[0]):
        data_split[i].at[j,"Date"]=dt.strptime(data_split[i].loc[j,"Date"], "%Y-%m-%d")
```

```python
data_split[0].head()
```

Tqdm library gives a progress plot and split the months into seasons

```python
from tqdm import tqdm
for i in range(0, len(data_split)):
    for j in tqdm(range(0,data_split[i]["Date"].shape[0])):
        k=data_split[i].loc[j,"Date"].strftime("%m")
        if k in ["12", "01", "02"]:#Dec,Jan,Feb is Summer
            data_split[i].at[j,"Season"]= "Summer"
        elif k in ["03", "04", "05"]:#Mar,Apr,May is Autumn
            data_split[i].at[j,"Season"]= "Autumn"
        elif k in ["06", "07", "08"]:#Jun,Jul,Aug is Winter
            data_split[i].at[j,"Season"]= "Winter"
        elif k in ["09", "10", "11"]:#Sep,Oct,Nov is Winter
            data_split[i].at[j,"Season"]= "Spring"
```

```python
data_split[0]
```

Replace the boolean string into integer 0 and 1

```python
for i in range(0, len(data_split)):
    data_split[i] = data_split[i].replace(to_replace = {"RainToday":{"No":0,"Yes":1}, "RainTomorrow":{"No":0,"Yes":1}})
```

One-hot encode season column to turn it into numerical value

```python
for i in range(0, len(data_split)):
    data_split[i]["Season"] = pd.Categorical(data_split[i]["Season"])
    season = pd.get_dummies(data_split[i]["Season"])
    data_split[i] = pd.concat([data_split[i],season],axis = 1)
```

```python
data_split[1].head(50)
```

Drop Date and Season since we have one-hot encoded the season

```python
data_seasonal=[]
for i in range(0, len(data_split)):
    data_seasonal.append(data_split[i].drop(["Season", "Date"], axis=1))
```

```python
data_seasonal[0].columns
```

## Exploratory Analysis

# Decision Tree Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
```

Splitting the training and test data

```python
data_x_train=[]
data_y_train=[]
data_x_test=[]
data_y_test=[]
for i in range(0, len(data_seasonal)):
    dx = pd.DataFrame(data_seasonal[i].drop(["RainTomorrow","Region"],axis = 1))
    dy = pd.DataFrame(data_seasonal[i]["RainTomorrow"])
    dxt, dxs, dyt, dys = train_test_split(dx, dy, test_size=0.25)
    data_x_train.append(dxt)
    data_y_train.append(dyt)
    data_x_test.append(dxs)
    data_y_test.append(dys)
```

```python
for i in range(0, len(data_seasonal)):
    pd.DataFrame(data_seasonal[i].drop(["RainTomorrow","Region"],axis = 1)).to_csv("DataReg/tab{}.csv".format(i))
```

```python
for i in range(0, len(data_x_train)):
    data_x_train[i].to_csv("DataReg/xtraindata{}.csv".format(i), index=False)
    data_x_test[i].to_csv("DataReg/xtestdata{}.csv".format(i), index=False)
    data_y_train[i].to_csv("DataReg/ytraindata{}.csv".format(i), index=False)
    data_y_test[i].to_csv("DataReg/ytestdata{}.csv".format(i), index=False)
```

Evaluating the model on each region

```python
decmod=[]
decsct=[]
decscs=[]
for i in range(0, len(data_x_train)):
    dectree = DecisionTreeClassifier(max_depth=3)
    dectree.fit(data_x_train[i], data_y_train[i])
    decsct.append(dectree.score(data_x_train[i], data_y_train[i]))
    decscs.append(dectree.score(data_x_test[i], data_y_test[i]))
    decmod.append(dectree)
```

```python
for i in range(0, len(decsct)):
    print("For the train data, score for model {} is: {}\n".format(i, decsct[i]))
    print("For the test data, score for model {} is: {}\n".format(i, decscs[i]))
```

```python
treedot = export_graphviz(decmod[0],                                      # the model
                          feature_names = data_x_train[0].columns,          # the features 
                          out_file = None,                              # output file
                          filled = True,                                # node colors
                          rounded = True,                               # make pretty
                          special_characters = True) 
graphviz.Source(treedot)
```

```python
for i in range(0,len(decmod)):
    h= open("DECModels/decmodel{}.sav", 'wb')
    pickle.dump(decmod[i], h)
```

```python
import tensorflow as tf
from tensorflow import keras
import pickle
tf.keras.backend.set_floatx('float64')
```

```python
from __future__ import absolute_import, division, print_function, unicode_literals
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
print(tf.config.experimental.list_physical_devices('GPU'))
```

```python
if tf.test.gpu_device_name():
    print("GPU yes")
```

```python
for i in tqdm(range(0, len(data_x_train))):
    dataset = tf.data.Dataset.from_tensor_slices((data_x_train[i].values,data_y_train[i]['RainTomorrow'].values))
    train_dataset = dataset.batch(1)
    model = keras.Sequential([
        keras.layers.Dense(100,activation = 'relu'),
        keras.layers.Dense(50,activation = 'relu'),
        keras.layers.Dense(1,activation = 'sigmoid')
    ])

    model.compile(
        optimizer = 'adam',
        loss = "binary_crossentropy",
        metrics = ['acc']
    )
    model.fit(train_dataset,epochs = 9,verbose = 2)
    model.save("kerasModelTrial{}.h5".format(i))
```

Defining the model with 4 hidden layers with 'relu' activation function

```python
for i in range(0, len(data_x_train)):
    model= keras.models.load_model("KerasModelTrial{}.h5".format(i))
    print("For model {} train: \n".format(i) + str(model.evaluate(x=data_x_train[i].values, y=data_y_train[i]['RainTomorrow'].values,verbose=0)))
    print("For model {} test: \n".format(i) + str(model.evaluate(x=data_x_test[i].values, y=data_y_test[i]['RainTomorrow'].values,verbose=0)))
```
