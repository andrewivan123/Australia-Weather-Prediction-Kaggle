```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
sb.set()
```

```python
data = pd.read_csv('weatherAUS.csv')
data.head()
```

```python
data.shape
```

```python
data_clean = data.dropna()
```

```python
data_clean.head()
```

```python
data_clean.Location.unique()
```

```python
reg= data_clean["Location"].drop_duplicates()
reg=reg.tolist()
reg
toregion={}
for i in range(0,len(reg)):
    if i==4 or i==23:
        toregion[reg[i]]="DROPPED"
    elif i<=7:
        toregion[reg[i]]="NSW"
    elif i<=14:
        toregion[reg[i]]="VIC"
    elif i<=17:
        toregion[reg[i]]="QLD"
    elif i<=20:
        toregion[reg[i]]="SA"
    elif i<=22:
        toregion[reg[i]]="WA"
    else:
        toregion[reg[i]]="NT"
        
```

```python
toregion
```

```python
data_clean["Region"]= data_clean["Location"].map(toregion)
```

```python
data_clean.head()
```

```python

f,axes = plt.subplots(1,1,figsize=(20,10))
sb.countplot(y=data_clean["Region"],)
```

Take Darwin since it has the most data

Drop wind direction and date column since logically it doesn't affect the chance of raining tomorrow

```python
data_clean = data_clean.drop(["RISK_MM","Location","WindGustDir","WindDir9am","WindDir3pm"],axis = 1)
```

```python
data_clean.head()
```

```python
p= data_clean["Region"].unique()
p =p.tolist()
```

```python
data_split=[]
for i in range(0,len(p)):
    data_split.append(data_clean[data_clean["Region"]==p[i]])
```

```python
del data_split[1]
```

```python
for i in range(0, len(data_split)):
    data_split[i].sort_values(by="Date")
    data_split[i]=data_split[i].reset_index(drop=True)
    for j in range(0, data_split[i].shape[0]):
        data_split[i].at[j,"Date"]=dt.strptime(data_split[i].loc[j,"Date"], "%Y-%m-%d")
```

```python
data_split[0].head()
```

```python
from tqdm import tqdm
for i in range(0, len(data_split)):
    for j in tqdm(range(0,data_split[i]["Date"].shape[0])):
        k=data_split[i].loc[j,"Date"].strftime("%m")
        if k in ["12", "01", "02"]:
            data_split[i].at[j,"Season"]= "Summer"
        elif k in ["03", "04", "05"]:
            data_split[i].at[j,"Season"]= "Autumn"
        elif k in ["06", "07", "08"]:
            data_split[i].at[j,"Season"]= "Winter"
        elif k in ["09", "10", "11"]:
            data_split[i].at[j,"Season"]= "Spring"
```

```python
data_split[0]
```

```python
for i in range(0, len(data_split)):
    data_split[i] = data_split[i].replace(to_replace = {"RainToday":{"No":0,"Yes":1}, "RainTomorrow":{"No":0,"Yes":1}})
```

```python
for i in range(0, len(data_split)):
    for j in range(0, data_split[i].shape[0]):
        if data_split[i].loc[j, "Season"]=="Summer":
            data_split[i].at[j, "Summer"]=1
            data_split[i].at[j, "Spring"]=0
            data_split[i].at[j, "Autumn"]=0
            data_split[i].at[j, "Winter"]=0
        elif data_split[i].loc[j, "Season"]=="Spring":
            data_split[i].at[j, "Summer"]=0
            data_split[i].at[j, "Spring"]=1
            data_split[i].at[j, "Autumn"]=0
            data_split[i].at[j, "Winter"]=0
        elif data_split[i].loc[j, "Season"]=="Autumn":
            data_split[i].at[j, "Summer"]=0
            data_split[i].at[j, "Spring"]=0
            data_split[i].at[j, "Autumn"]=1
            data_split[i].at[j, "Winter"]=0
        elif data_split[i].loc[j, "Season"]=="Winter":
            data_split[i].at[j, "Summer"]=0
            data_split[i].at[j, "Spring"]=0
            data_split[i].at[j, "Autumn"]=0
            data_split[i].at[j, "Winter"]=1
```

```python
data_seasonal=[]
for i in range(0, len(data_split)):
    data_seasonal.append(data_split[i].drop(["Season", "Date"], axis=1))
```

```python
data_seasonal[0].columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
```

```python
data_x_train=[]
data_y_train=[]
data_x_test=[]
data_y_test=[]
for i in range(0, len(data_seasonal)):
    dx = pd.DataFrame(data_seasonal[i].drop(["RainTomorrow","Region"],axis = 1))
    dy = pd.DataFrame(data_seasonal[i]["RainTomorrow"])
    dxt, dxs, dyt, dys = train_test_split(dx, dy, test_size=0.25)
    data_x_train.append(dxt)
    data_y_train.append(dyt)
    data_x_test.append(dxs)
    data_y_test.append(dys)
```

```python
decmod=[]
decsct=[]
decscs=[]
for i in range(0, len(data_x_train)):
    dectree = DecisionTreeClassifier(max_depth=3)
    dectree.fit(data_x_train[i], data_y_train[i])
    decsct.append(dectree.score(data_x_train[i], data_y_train[i]))
    decscs.append(dectree.score(data_x_test[i], data_y_test[i]))
    decmod.append(dectree)
```

```python
for i in range(0, len(decsct)):
    print("For the train data, score for model {} is: {}\n".format(i, decsct[i]))
    print("For the test data, score for model {} is: {}\n".format(i, decscs[i]))
```

```python
treedot = export_graphviz(decmod[0],                                      # the model
                          feature_names = data_x_train[0].columns,          # the features 
                          out_file = None,                              # output file
                          filled = True,                                # node colors
                          rounded = True,                               # make pretty
                          special_characters = True) 
graphviz.Source(treedot)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmod=[]
logsct=[]
logscs=[]
for i in range(0, len(data_x_train)):
    logreg=LogisticRegression(solver="lbfgs")
    logreg.fit(data_x_train[i], data_y_train[i])
    logsct.append(logreg.score(data_x_train[i], data_y_train[i]))
    logscs.append(logreg.score(data_x_test[i], data_y_test[i]))
    logmod.append(logreg)
```

```python
for i in range(0, len(logsct)):
    print("For the train data, score for model {} is: {}\n".format(i, logsct[i]))
    print("For the test data, score for model {} is: {}\n".format(i, logscs[i]))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcmod=[]
rfcsct=[]
rfcscs=[]
for i in range(0, len(data_x_train)):
    rfc = RandomForestClassifier(n_estimators=100,bootstrap=True)
    rfc.fit(data_x_train[i], data_y_train[i])
    rfcsct.append(rfc.score(data_x_train[i], data_y_train[i]))
    rfcscs.append(rfc.score(data_x_test[i], data_y_test[i]))
    rfcmod.append(rfc)
```

```python
for i in range(0, len(rfcsct)):
    print("For the train data, score for model {} is: {}\n".format(i, rfcsct[i]))
    print("For the test data, score for model {} is: {}\n".format(i, rfcscs[i]))
```
