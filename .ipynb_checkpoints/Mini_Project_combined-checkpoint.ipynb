```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as dt
import pickle
sb.set()
```

```python
data = pd.read_csv('weatherAUS.csv')
data.head()
```

```python
data.columns
```

# Data Cleaning and Preprocessing

Dropping rows with NaN values

```python
data_clean = data.dropna()
```

```python
data_clean.head()
```

## Grouping cities based on their state

```python
data_clean.Location.unique()
```

Mapping each city to their respective state

```python
reg= data_clean["Location"].drop_duplicates()
reg=reg.tolist()
toregion={}
for i in range(0,len(reg)):
    if i==4 or i==23:
        toregion[reg[i]]="DROPPED"
    elif i<=7:
        toregion[reg[i]]="NSW"
    elif i<=14:
        toregion[reg[i]]="VIC"
    elif i<=17:
        toregion[reg[i]]="QLD"
    elif i<=20:
        toregion[reg[i]]="SA"
    elif i<=22:
        toregion[reg[i]]="WA"
    else:
        toregion[reg[i]]="NT"
        
```

```python
toregion
```

Adding a new feature column called "Region" which represents the city's state

```python
data_clean["Region"]= data_clean["Location"].map(toregion)
```

```python
data_clean.head()
```

```python
f,axes = plt.subplots(1,1,figsize=(20,10))
sb.countplot(y=data_clean["Region"],)
```

Drop wind direction, location and risk_mm

```python
data_clean = data_clean.drop(["RISK_MM","Location","WindGustDir","WindDir9am","WindDir3pm"],axis = 1)
```

```python
data_clean.head()
```

```python
p = data_clean["Region"].unique()
p = p.tolist()
```

```python
data_split=[]
for i in range(0,len(p)):
    data_split.append(data_clean[data_clean["Region"]==p[i]])
```

Delete "DROPPED" Region

```python
del data_split[1]
```

## Converting dates into seasons

Convert string in "Date" to date-time data type

```python
for i in range(0, len(data_split)):
    data_split[i].sort_values(by="Date")
    data_split[i]=data_split[i].reset_index(drop=True)
    for j in range(0, data_split[i].shape[0]):
        data_split[i].at[j,"Date"]=dt.strptime(data_split[i].loc[j,"Date"], "%Y-%m-%d")
```

```python
data_split[0].head()
```

Tqdm library gives a progress plot and split the months into seasons

```python
from tqdm import tqdm
for i in range(0, len(data_split)):
    for j in tqdm(range(0,data_split[i]["Date"].shape[0])):
        k=data_split[i].loc[j,"Date"].strftime("%m")
        if k in ["12", "01", "02"]:#Dec,Jan,Feb is Summer
            data_split[i].at[j,"Season"]= "Summer"
        elif k in ["03", "04", "05"]:#Mar,Apr,May is Autumn
            data_split[i].at[j,"Season"]= "Autumn"
        elif k in ["06", "07", "08"]:#Jun,Jul,Aug is Winter
            data_split[i].at[j,"Season"]= "Winter"
        elif k in ["09", "10", "11"]:#Sep,Oct,Nov is Winter
            data_split[i].at[j,"Season"]= "Spring"
```

```python
data_split[0]
```

Replace the boolean string into integer 0 and 1

```python
for i in range(0, len(data_split)):
    data_split[i] = data_split[i].replace(to_replace = {"RainToday":{"No":0,"Yes":1}, "RainTomorrow":{"No":0,"Yes":1}})
```

One-hot encode season column to turn it into numerical value

```python
for i in range(0, len(data_split)):
    data_split[i]["Season"] = pd.Categorical(data_split[i]["Season"])
    season = pd.get_dummies(data_split[i]["Season"])
    data_split[i] = pd.concat([data_split[i],season],axis = 1)
```

```python
data_split[1].head(50)
```

Drop Date and Season since we have one-hot encoded the season

```python
data_seasonal=[]
for i in range(0, len(data_split)):
    data_seasonal.append(data_split[i].drop(["Season", "Date"], axis=1))
```

```python
data_seasonal[0].columns
```

```python
data_combined = pd.concat(data_seasonal)

data_combined["Region"] = pd.Categorical(data_combined["Region"])
region = pd.get_dummies(data_combined["Region"])
data_combined = pd.concat([data_combined,region],axis = 1)
data_combined
```

## Exploratory Analysis

```python
f, axes = plt.subplots(17, 3, figsize=(18, 72))
colors = ["r", "g", "b", "m", "c", "y",]

count = 0
for var in data_combined:
    if count<=16:
        sb.boxplot(data_combined[var], orient = "h", color = colors[(count%6)], ax = axes[count,0])
        sb.distplot(data_combined[var], color = colors[(count%6)], ax = axes[count,1])
        sb.violinplot(data_combined[var], color = colors[(count%6)], ax = axes[count,2])
    count += 1
```

# Decision Tree Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
from sklearn.metrics import confusion_matrix
from os import system
```

Splitting the training and test data

```python
data_x = data_combined.drop(["RainTomorrow","Region"],axis = 1)
data_y = data_combined["RainTomorrow"]

data_x_train, data_x_test, data_y_train, data_y_test = train_test_split(data_x, data_y, test_size=0.25,random_state=42)
print(data_x_train.__len__(), data_y_train.__len__())
```

```python
data_x_train.to_csv("DataCom/xtraindata.csv", index=False)
data_x_test.to_csv("DataCom/xtestdata.csv", index=False)
data_y_train.to_csv("DataCom/ytraindata.csv", index=False)
data_y_test.to_csv("DataCom/ytestdata.csv", index=False)
```

Fitting the model and evaluating the model on each region

```python
dectree = DecisionTreeClassifier(max_depth=5)
model = dectree.fit(data_x_train, data_y_train)
print("For the train data, score for model is: {}\n".format(dectree.score(data_x_train, data_y_train)))
print("For the test data, score for model is: {}\n".format(dectree.score(data_x_test, data_y_test)))
```

```python
treedot = export_graphviz(model,                                      # the model
                          feature_names = data_x_train.columns,          # the features 
                          out_file = None,                              # output file
                          filled = True,                                # node colors
                          rounded = True,                               # make pretty
                          special_characters = True) 
graphviz.Source(treedot)
```

```python
pickle.dump(dectree, open("DECModels/decmodelcombined.sav",'wb'))
```

## Confusion Matrix

```python
data_x_test_pred = dectree.predict(data_x_test)

sb.heatmap(confusion_matrix(data_y_test, data_x_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18},cmap = "BuPu")
```

## Neural Network using Tensorflow

```python
import tensorflow as tf
from tensorflow import keras
```

```python
dataset = tf.data.Dataset.from_tensor_slices((data_x_train.values,data_y_train.values))
```

```python
train_dataset = dataset.batch(256)
```

```python
train_dataset
```

Defining the model with 2 hidden layers with 'relu' activation function

```python
model = keras.Sequential([
    keras.layers.Dense(100,activation = 'relu'),
    keras.layers.Dense(50,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = "binary_crossentropy",
    metrics = ['acc']
)
```

```python
model.fit(train_dataset,epochs = 2048,verbose = 2)
```

```python
model.save("kerasmodelcombined.h5")
```

```python
model.evaluate(data_x_train.values,data_y_train.values)
```

```python
model.evaluate(data_x_test.values,data_y_test.values)
```
